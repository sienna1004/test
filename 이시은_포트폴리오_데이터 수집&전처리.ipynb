#### 국내 주요 음료 유통업체 공식몰 웹크롤링: 음료 리스트 추출

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time

# 웹 드라이버 설정
driver = webdriver.Chrome()

# URL 리스트 정의
urls = [
    'https://brand.naver.com/binggrae/category/d405291895da4ec4822a3c7cbd9efbe8?cp=1',
    'https://brand.naver.com/binggrae/category/58af36386a594c419bbe87ccd4f18a74?cp=1',
    'https://brand.naver.com/binggrae/category/cb471bcd275648b1857377f4b49a73e5?cp=1',
    'https://brand.naver.com/binggrae/category/dc667c6583c14e0e8ac6397c01c9c815?cp=1'
]

# 모든 페이지의 시리즈 이름을 저장할 리스트
all_bv_names = []

# 각 URL에 대해 반복
for url in urls:
    driver.get(url)
    time.sleep(2)  # 페이지 로드 대기

    # 페이지 내리기
    for i in range(3):
        driver.find_element(By.TAG_NAME, value="body").send_keys(Keys.PAGE_DOWN)  # 내리기
        time.sleep(1)

    # Selenium을 이용해 현재 페이지의 HTML을 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # 음료 이름 가져오기
    bv_list = soup.find_all('strong', {'class': '_3pA0Duwrhw'})
    for bv in bv_list:
        # 텍스트를 추출하고 공백 제거
        all_bv_names.append(bv.get_text().strip())

# 결과 출력
for bv_name in all_bv_names:
    print(bv_name)

# DataFrame에 저장하고 엑셀 파일로 내보내기
df_빙그레_bv = pd.DataFrame(all_bv_names, columns=["Product Name"])

# 브라우저 닫기
driver.quit()
```

#### 가공식품 영양성분 DB 매핑을 통한 정보 추출

```python
import pandas as pd

# 데이터 불러오기
bv_list_df = pd.read_excel('음료리스트_11.15_v1.0_음료 리스트 1차 완료.xlsx')
nutrition_db_df = pd.read_excel('★ 20241016_ 영양성분DB_가공식품DB.xlsx')

# '식품명'과 '식품중량'이 중복되는 경우 첫 번째 값만 남기고 제거
nutrition_db_df = nutrition_db_df.drop_duplicates(subset=['식품명', '식품중량'])

# '음료명', '식품명', '용량', '식품중량'에서 공백 제거
bv_list_df['음료명'] = bv_list_df['음료명'].str.strip().str.replace(' ', '', regex=False)
bv_list_df['용량'] = bv_list_df['용량'].astype(str).str.strip().str.replace(' ', '', regex=False)

nutrition_db_df['식품명'] = nutrition_db_df['식품명'].str.strip().str.replace(' ', '', regex=False)
nutrition_db_df['식품중량'] = nutrition_db_df['식품중량'].astype(str).str.strip().str.replace(' ', '', regex=False)

# '음료명'과 '식품명', '용량'과 '식품중량'이 일치하는 데이터를 기준으로 병합
merged_df = bv_list_df.merge(
    nutrition_db_df[['식품명', '식품중량', '대표식품명', '식품중분류명', '식품소분류명', '영양성분함량기준량', '에너지(kcal)', '단백질(g)', '지방(g)', '탄수화물(g)', '당류(g)', '나트륨(mg)', '콜레스테롤(mg)', '포화지방산(g)', '트랜스지방산(g)', '알룰로오스(g)','에리스리톨(g)','품목제조보고번호', '제조사명', '유통업체명']],
    how='left',
    left_on=['음료명', '용량'],
    right_on=['식품명', '식품중량']
)

# 불필요한 열 제거 (식품명, 식품중량 등)
merged_df = merged_df.drop(columns=['식품명', '식품중량'])

# 결과 저장
merged_df.to_excel('음료리스트_11.16_v1.0_영양성분DB추가.xlsx', index=False)
```

#### 영양성분 DB 공백 데이터: 성분표 이미지 캡처 후 네이버 OCR API 활용 데이터 추출

```python
import os
import requests
import pandas as pd
import re
from uuid import uuid4
from time import time

def extract_nutrition(text):
    """영양성분 정보 추출"""
    if not text:
        return {}
    
    patterns = {
        '영양성분함량기준량': r'(\d+\.?\d*)\s*[mL리터]',
        '에너지(kcal)': r'(\d+)\s*[kK][cC]al',
        '단백질(g)': r'단백질\s*[:：]?\s*(\d+\.?\d*)\s*[gG]',
        '지방(g)': r'지방\s*[:：]?\s*(\d+\.?\d*)\s*[gG]',
        '탄수화물(g)': r'탄수화물\s*[:：]?\s*(\d+\.?\d*)\s*[gG]',
        '당류(g)': r'당류\s*[:：]?\s*(\d+\.?\d*)\s*[gG]',
        '나트륨(mg)': r'나트륨\s*[:：]?\s*(\d+\.?\d*)\s*[mM][gG]',
        '품목제조보고번호': r'품목[^\d]*(\d+)'
    }
    
    return {k: float(re.search(v, text).group(1)) if re.search(v, text) else None 
            for k, v in patterns.items()}

def ocr_extract(image_path, api_url, secret_key):
    """네이버 OCR API로 텍스트 추출"""
    if not os.path.exists(image_path):
        return None
        
    request_json = {
        'images': [{'format': 'jpg', 'name': 'demo'}],
        'requestId': str(uuid4()),
        'version': 'V2',
        'timestamp': int(time() * 1000)
    }
    
    try:
        with open(image_path, 'rb') as file:
            response = requests.post(
                api_url,
                headers={'X-OCR-SECRET': secret_key},
                data={'message': json.dumps(request_json).encode('UTF-8')},
                files=[('file', file)]
            )
        
        return ' '.join([field['inferText'] for field in response.json()['images'][0]['fields']]) \
               if response.status_code == 200 else None
    except Exception as e:
        print(f"OCR 오류: {e}")
        return None

def update_excel(excel_path, image_folder, api_url, secret_key):
    """엑셀 파일 영양성분 정보 업데이트"""
    df = pd.read_excel(excel_path)
    
    for img in os.listdir(image_folder):
        if img.lower().endswith('.png'):
            beverage_name = os.path.splitext(img)[0]
            text = ocr_extract(os.path.join(image_folder, img), api_url, secret_key)
            
            if text and (mask := df['음료명'] == beverage_name).any():
                nutrition_data = extract_nutrition(text)
                for column, value in nutrition_data.items():
                    if column in df.columns and value is not None:
                        df.loc[mask, column] = value
    
    df.to_excel("음료리스트_updated.xlsx", index=False)

if __name__ == "__main__":
    API_URL = 'NAVER OCR API URL'
    SECRET_KEY = 'API SECRET KEY'
    update_excel("음료리스트.xlsx", "음료_영양성분 캡쳐", API_URL, SECRET_KEY)
```

#### 기존 DB의 품목제조보고번호를 활용한 openAPI 연계 원재료 데이터 추출

```python
import pandas as pd
import requests
import time
from urllib.parse import quote

def get_ingredients(api_key, param_key, value):
    """API를 통해 원재료 정보 조회"""
    try:
        value_str = str(value).strip() if pd.notna(value) else ''
        url = f"https://openapi.foodsafetykorea.go.kr/api/{api_key}/C002/json/1/1000/{param_key}={quote(value_str)}"
        
        response = requests.get(url).json()
        if 'C002' in response and int(response['C002']['total_count']) > 0:
            return response['C002']['row'][0]['RAWMTRL_NM']
        return None
    except Exception as e:
        print(f"API 오류: {e}")
        return None

def update_excel(file_path, api_key):
    """엑셀 파일의 원재료 정보 업데이트"""
    df = pd.read_excel(file_path, dtype={'품목제조보고번호': str})
    df['원재료'] = ''
    
    for idx, row in df.iterrows():
        ingredients = (
            get_ingredients(api_key, 'PRDLST_REPORT_NO', row.get('품목제조보고번호')) or 
            get_ingredients(api_key, 'PRDLST_NM', row.get('음료명'))
        )
        
        if ingredients:
            df.at[idx, '원재료'] = ingredients
            print(f"행 {idx+2}: {row['음료명']} - 업데이트 완료")
        
        time.sleep(0.5)
    
    output_file = file_path.replace('.xlsx', '_updated.xlsx')
    df.to_excel(output_file, index=False)

if __name__ == "__main__":
    API_KEY = 'API KEY'
    update_excel('음료리스트_11.21_v6.0_ 캡쳐 내용 업데이트.xlsx', API_KEY)
```

### 2.1.7 '100ml당 가격' 열 정규화

```python
# 1단계: 10ml당 가격을 100ml당 가격으로 변환
def convert_to_100ml(price):
    if '10ml당' in price:
        # 괄호와 '원' 제거, '10ml당' 제거 후 숫자만 추출
        num = price.replace('(', '').replace(')', '').replace('10ml당 ', '').replace('원', '').replace(',', '')
        # 10을 곱해서 100ml당 가격으로 변환
        return f'(100ml당 {int(num)*10}원)'
    return price

df_beverage4['100ml당 가격'] = df_beverage4['100ml당 가격'].apply(convert_to_100ml)

# 2단계: 괄호 제거
df_beverage4['100ml당 가격'] = df_beverage4['100ml당 가격'].str.replace('(', '').str.replace(')', '')

# 3단계: '100ml당 ' 제거
df_beverage4['100ml당 가격'] = df_beverage4['100ml당 가격'].str.replace('100ml당 ', '')

# 4단계: '원' 제거
df_beverage4['100ml당 가격'] = df_beverage4['100ml당 가격'].str.replace('원', '')

# 5단계: 쉼표 제거
df_beverage4['100ml당 가격'] = df_beverage4['100ml당 가격'].str.replace(',', '')

# 6단계: 숫자로 변환
df_beverage4['100ml당 가격'] = df_beverage4['100ml당 가격'].astype(int)
```

### 2.1.17 '인기도'파생 변수 생성

- '인기도'에 사용된 변수: 평점, 24년도 리뷰수, 월간 검색량, 월간 콘텐츠 발행량
- '평점'과 '24년도 리뷰수'로 '리뷰 신뢰도 점수' 파생변수 생성
- 각 지표 정규화 후, 각각 가중치 적용하여 '인기도' 파생 변수 생성 ( 리뷰 신뢰도 점수 50%, 검색량 30%, 콘텐츠량 20%)

```python
df_beverage11 = df_beverage10.copy()

# 리뷰 신뢰도 점수 생성
df_beverage11['리뷰 신뢰도 점수'] = df_beverage11['평점'] * df_beverage11['24년도 리뷰수']

# 정규화 함수 정의 (0-100점 스케일로)
def min_max_scale(x):
    return 100 * (x - x.min()) / (x.max() - x.min())

# 각 지표 정규화
df_beverage11['리뷰 신뢰도 정규화'] = min_max_scale(df_beverage11['리뷰 신뢰도 점수'])
df_beverage11['검색량 정규화'] = min_max_scale(df_beverage11['월간_검색량'])
df_beverage11['콘텐츠량 정규화'] = min_max_scale(df_beverage11['월간_콘텐츠_발행량'])

# 인기도 점수 계산 (가중치 적용)
df_beverage11['인기도'] = (df_beverage11['리뷰 신뢰도 정규화'] * 0.5 + 
               df_beverage11['검색량 정규화'] * 0.3 + 
               df_beverage11['콘텐츠량 정규화'] * 0.2)
```

- 이러한 분포 불균형은 데이터 내 다수의 극단값에서 기인한 것으로 판단되어, 로그 변환과 Winsorization 기법을 적용해 상하위 5%의 극단값을 처리한 후 인기도 점수를 재계산하기로 결정함

### 2.1.18 로그변환 + Winsorization

```python
import numpy as np

df_beverage13 = df_beverage10.copy()

# 리뷰 신뢰도 점수 생성
df_beverage13['리뷰_신뢰도_점수'] = df_beverage13['평점'] * df_beverage13['24년도 리뷰수']

# 로그 변환 함수 정의 (log1p 사용하여 0 처리)
def log_transform(series):
    return np.log1p(series)

# 각 지표에 로그 변환 적용
df_beverage13['리뷰_신뢰도_점수_log'] = log_transform(df_beverage13['리뷰_신뢰도_점수'])
df_beverage13['검색량_log'] = log_transform(df_beverage13['월간_검색량'])
df_beverage13['콘텐츠량_log'] = log_transform(df_beverage13['월간_콘텐츠_발행량'])

# Winsorization 함수 정의 (상하위 5% 기준)
def winsorize_series(series):
    lower = np.percentile(series, 5)
    upper = np.percentile(series, 95)
    return pd.Series([min(max(x, lower), upper) for x in series])

# 로그 변환된 지표들에 Winsorization 적용
df_beverage13['리뷰_신뢰도_점수_log_winsorized'] = winsorize_series(df_beverage13['리뷰_신뢰도_점수_log'])
df_beverage13['검색량_log_winsorized'] = winsorize_series(df_beverage13['검색량_log'])
df_beverage13['콘텐츠량_log_winsorized'] = winsorize_series(df_beverage13['콘텐츠량_log'])

# 정규화 함수 정의 (0-100점 스케일로)
def min_max_scale(x):
    if x.max() == x.min():
        return x * 0
    return 100 * (x - x.min()) / (x.max() - x.min())

# Winsorization된 로그 변환 지표들을 정규화
df_beverage13['리뷰_신뢰도_정규화'] = min_max_scale(df_beverage13['리뷰_신뢰도_점수_log_winsorized'])
df_beverage13['검색량_정규화'] = min_max_scale(df_beverage13['검색량_log_winsorized'])
df_beverage13['콘텐츠량_정규화'] = min_max_scale(df_beverage13['콘텐츠량_log_winsorized'])

# 인기도 점수 계산 (가중치 적용)
df_beverage13['인기도'] = (df_beverage13['리뷰_신뢰도_정규화'] * 0.5 + 
               df_beverage13['검색량_정규화'] * 0.3 + 
               df_beverage13['콘텐츠량_정규화'] * 0.2)

# 원본 방식으로도 계산 (비교용)
df_beverage13['리뷰_신뢰도_정규화_원본'] = min_max_scale(df_beverage13['리뷰_신뢰도_점수'])
df_beverage13['검색량_정규화_원본'] = min_max_scale(df_beverage13['월간_검색량'])
df_beverage13['콘텐츠량_정규화_원본'] = min_max_scale(df_beverage13['월간_콘텐츠_발행량'])

df_beverage13['인기도_원본'] = (df_beverage13['리뷰_신뢰도_정규화_원본'] * 0.5 + 
                    df_beverage13['검색량_정규화_원본'] * 0.3 + 
                    df_beverage13['콘텐츠량_정규화_원본'] * 0.2)

# 데이터 변환 과정 비교
print("=== 데이터 변환 과정 비교 ===")
comparison_cols = [
    '리뷰_신뢰도_점수', '리뷰_신뢰도_점수_log', '리뷰_신뢰도_점수_log_winsorized',
    '월간_검색량', '검색량_log', '검색량_log_winsorized',
    '월간_콘텐츠_발행량', '콘텐츠량_log', '콘텐츠량_log_winsorized'
]
print("\n각 단계별 데이터 통계량:")
print(df_beverage13[comparison_cols].describe())

# 상위 50개 음료 확인
print("\n=== 상위 50개 음료 ===")
top_50 = df_beverage13.nlargest(50, '인기도')[['음료명', '인기도', '평점', '24년도 리뷰수','리뷰수', '월간_검색량', '월간_콘텐츠_발행량']]
print(top_50.to_string())

# 기초 통계량 확인
print("\n=== 인기도 점수 기초 통계량 ===")
print(df_beverage13['인기도'].describe())

# 상위 10개 음료의 세부 정보 확인
print("\n=== 상위 10개 음료의 정규화된 점수 ===")
detail_cols = ['음료명', '리뷰_신뢰도_정규화', '검색량_정규화', '콘텐츠량_정규화', '인기도']
top_10_details = df_beverage13.nlargest(10, '인기도')[detail_cols]
print(top_10_details.to_string())
```

- 로그변환을 통해 극단값들의 스케일이 효과적으로 감소됨
- 로그 변환된 데이터에 Winsorization을 추가 적용함으로써 보다 안정적인 분포 형성
- 결과적으로 인기도 점수 분포가 균일해지고 상하위 간 점수 격차가 적절히 완화됨
- 이와 같은 처리로 이전 대비 더욱 안정적이고 현실적인 인기도 평가가 가능해짐

### 2.1.22 대체당 추가 기재

```python
# 대체당 원재료 추가할 음료 목록
sweetener_dict = {
    '닥터페퍼 제로': '수크랄로스, 아스파탐, 아세설팜칼륨',
    '스프라이트 제로': '아세설팜칼륨, 수크랄로스, 스테비아',
    '캐나다 드라이 진저에일 토닉 제로': '수크랄로스, 아세설팜칼륨',
    '코카-콜라 제로': '아세설팜칼륨, 수크랄로스',
    '코카-콜라 제로 레몬​': '수크랄로스, 아세설팜칼륨',
    '코카-콜라 제로 체리': '수크랄로스, 아세설팜칼륨',
    '코카-콜라 제로제로': '수크랄로스, 아세설팜칼륨',
    '토레타 제로': '아세설팜칼륨, 수크랄로스',
    '파워에이드 제로': '수크랄로스, 아세설팜칼륨',
    '환타 제로 오렌지향': '수크랄로스, 아세설팜칼륨',
    '환타 제로 파인애플향': '수크랄로스, 아세설팜칼륨',
    '환타 제로 포도향': '수크랄로스, 아세설팜칼륨',
    '갈배사이다 제로': '수크랄로스, 아세설팜칼륨',
    '토레타 제로 라벨프리': '아세설팜칼륨, 수크랄로스'
}

# 각 음료의 원재료 업데이트
for beverage, sweeteners in sweetener_dict.items():
    mask = df_beverage15['음료명'] == beverage
    if mask.any():  # 해당 제품명이 존재하는 경우에만 업데이트
        df_beverage15.loc[mask, '원재료'] = df_beverage15.loc[mask, '원재료'] + ', ' + sweeteners
    else:
        print(f"경고: '{beverage}' 제품을 찾을 수 없습니다.")

# 업데이트된 데이터프레임 확인
print("\n=== 원재료 업데이트 결과 ===")
for beverage in sweetener_dict.keys():
    mask = df_beverage15['음료명'] == beverage
    if mask.any():
        updated_ingredients = df_beverage15.loc[mask, '원재료'].values[0]
        print(f"{beverage}: {updated_ingredients}")
    else:
        print(f"'{beverage}' 제품을 찾을 수 없음")
```

### 2.2.2 '한달사용기','재구매'열 추가

```python
# '한달사용기' 열 추가 ('한달사용기' 또는 '한달사용'으로 시작하는 경우 1)
df_review_1['한달사용기'] = (df_review_1['리뷰 내용'].str.startswith(('한달사용기', '한달사용'))).astype(int)

# 텍스트 제거 ('한달사용기'와 '한달사용' 각각의 경우 처리)
mask_usage1 = df_review_1['리뷰 내용'].str.startswith('한달사용기')
mask_usage2 = df_review_1['리뷰 내용'].str.startswith('한달사용')

# replace 메서드에서 n=1 파라미터는 첫 번째로 발견된 텍스트만 대체하도록 합니다
df_review_1.loc[mask_usage1, '리뷰 내용'] = df_review_1['리뷰 내용'].str.replace('한달사용기', '', n=1)
df_review_1.loc[mask_usage2, '리뷰 내용'] = df_review_1['리뷰 내용'].str.replace('한달사용', '', n=1)

# '재구매' 열 추가 ('재구매'로 시작하는 경우 1)
df_review_1['재구매'] = (df_review_1['리뷰 내용'].str.startswith('재구매')).astype(int)

# '재구매' 텍스트 제거 
mask_repurchase = df_review_1['리뷰 내용'].str.startswith('재구매')
df_review_1.loc[mask_repurchase, '리뷰 내용'] = df_review_1['리뷰 내용'].str.replace('재구매', '', n=1)
```
